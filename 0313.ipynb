## 단어 토큰화

```python
from nltk import word_tokenize

sentence = '나는 아침에 웹소설과 웹툰을 읽어야만 정신을 차릴 수가 있습니다.'

words = word_tokenize(sentence)
print(words)
```

```python
# 문서를 받아서 문장 단위로 단어 토큰화를 수행해서 리턴하는 함수
def tokenize_text (text) -> list :
    sentences = sent_tokenize(text)
    word_tokens = [word_tokenize(sentence) for sentence in sentences]
    return word_tokens
```

```python
test_text = '''
나는 어제 저녁을 먹은 후 침대에 누워서 크라임씬을 봤다. 그러다가 잠들었다. 
심지어 휴대폰은 충전도 안해둬서 아침에 폰의 전원이 꺼져있었다.
그래서 알람을 듣지 못하고 늦잠을 자버렸다.
7시에 일어나서 부랴부랴 준비를 하고 집을 나섰다.
다행히 지각은 하지 않았다!
'''

word_tokens = tokenize_text(test_text)
print(word_tokens)
```

## 희소 행렬 표현 방식

```python
# COO 방식 - 0 이 아닌 데이터의 좌표를 기록하는 방식
dense = np.array([[3, 0, 1], [0, 2, 0]])

# 0이 아닌 데이터의 가중치와 행과 열번호를 기록

from scipy import sparse
sparse_coo = sparse.coo_matrix(([3, 1, 2], (np.array([0, 0, 1]), np.array([0, 2, 1]))))
print(sparse_coo)

# 복원
print(sparse_coo.toarray())
```

```python
# CSR 방식 - 반복적으로 등장하는 위치를 한 번만 표현하기 위해서 시작 위치를 가리키는 방식
dense = np.array(
[[0, 0, 1, 0, 0, 5],
[1, 4, 0, 3, 2, 5],
[0, 6, 0, 3, 0, 0],
[2, 0, 0, 0, 0, 0],
[0, 0, 0, 7, 0, 8],
[1, 0, 0, 0, 0, 0]] )

print(dense)

# 0 이 아닌 데이터 추출 - 데이터 배열
data = np.array([1, 5, 1, 4, 3, 2, 5, 6, 3, 2, 7, 8, 1])

# 0이 아닌 데이터의 행 위치 - 행 인덱스
row_pos = np.array([0, 0, 1, 1, 1, 1, 1, 2, 2, 3, 4, 4, 5])

# 0이 아닌 데이터의 열 위치- 열 인덱스
col_pos = np.array([2, 5, 0, 1, 3, 4, 5, 1, 3, 0, 3, 5, 0])

# 행 인덱스가 중복되는 경우가 많음
# 중복된 행 인덱스를 1번씩만 표현
# 각 행의 시작 인덱스 위치를 기억하고 마지막에는 0이 아닌 데이터의 개수를 설정
row_pos_ind = np.array([0, 2, 7, 9, 10, 12, 13])

# csr 방식으로 지정
sparse_csr = sparse.csr_matrix((data, col_pos, row_pos_ind))
print(sparse_csr) # 희소 행렬로 출력
print(sparse_csr.toarray()) # 밀집 행렬로 출력
```

### sklearn API 를 이용한 피처 벡터화

```python
# 데이터 생성
corpus = ['나는 어제 저녁을 먹은 후 침대에 누워서 휴대폰을 봤다.',
'저녁을 먹고 나서 잠들었다.',
'심지어 저녁에 휴대폰 충전도 안해둬서 아침에 휴대폰 전원이 꺼져있었다.',
'나는 그래서 알람을 듣지 못하고 아침에 늦잠을 자버렸다.',
'7시에 일어나서 빨리 준비를 하고 휴대폰을 챙겨서 집을 나섰다.',
'다행히 빨리 준비해서 다행히 지각은 하지 않았다!']
print(corpus)
```

```python
# CountVectorizer : 등장 횟수만으로 가중치를 표현
from sklearn.feature_extraction.text import CountVectorizer

cvect = CountVectorizer()
# 기존 문자열을 가지고 훈련
# 문서에 등장한 모든 단어를 indexing 만 수행
cvect.fit(corpus)

# corpus의 문장을 벡터화
# 앞의 숫자 0 행은 행번호이고 뒤의 숫자는 단어의 인덱스이고 세번째 숫자가 등장 횟수
dtm = cvect.transform(corpus)
print(dtm)
```

```python
# 각 단어 확인
print(cvect.vocabulary_)
```

```python
# 단어를 인덱스 순서대로 확인
vocab = cvect.get_feature_names_out()
print(vocab)
```

```python
# 문장들을 피처화
# 원핫 인코딩 과 다른 점은 하나의 행에 0이 아닌 값이 여러개 있고 0이나 1아닌 값이 존재
df_dtm = pd.DataFrame(dtm.toarray(), columns = vocab)
print(df_dtm)
```

```python
#TfidfVectorizer
# 문장 내에서 여러번 등장하면 가중치를 부여하고 여러 문장에서 등장하면 패털티를 부과
# 데이터 생성
corpus = ['나는 어제 저녁을 먹은 후 침대에 누워서 휴대폰을 봤다.',
'저녁을 먹고 나서 잠들었다.',
'심지어 저녁에 휴대폰 충전도 안해둬서 아침에 휴대폰 전원이 꺼져있었다.',
'나는 그래서 알람을 듣지 못하고 아침에 늦잠을 자버렸다.',
'7시에 일어나서 빨리 준비를 하고 휴대폰을 챙겨서 집을 나섰다.',
'다행히 빨리 준비해서 다행히 지각은 하지 않았다!']
print(corpus)
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer

tfidfv = TfidfVectorizer()
# 기존 문자열을 가지고 훈련
# 문서에 등장한 모든 단어를 indexing 만 수행
tfidfv.fit(corpus)
```

```python
# 각 단어 확인
print(tfidfv.vocabulary_)
```

```python
# 문장들을 피처화
vocab = tfidfv.get_feature_names_out()
df_dtm = pd.DataFrame(tfidfv.transform(corpus).toarray(), columns = vocab)
print(df_dtm)
```

```python
# 읽어올 데이터의 개수 입력 받기
# 입력 받은 숫자를 가지고 읽어올 페이지의 개수를 생성
# 한페이지에 기사는 10개

page_num = int(int(input('읽어올 데이터의 개수 :')) / 10 + 1) # 파이썬은 정수 나누기 정수를 해도 float이다.
print(page_num)
```

```python
# 파싱한 내용을 저장할 텍스트 파일 생성
output_file = open(keyword + '.txt', 'w', encoding = 'utf8')

# 기사 링크를 한 곳에 저장하기
for i in range(page_num) :
    # 페이지 당 기사가 10개
    current_page_num = 1+  i * 10
    target_url = "https://www.donga.com/news/search?p=" + str(current_page_num) \
    + "&query=" + quote(keyword) \
    + "&check_news=91&sorting=1&search_date=1&v1=&v2=&more=1"
    print(target_url)
    source_code_from_URL = requests.get(target_url)
    bs = BeautifulSoup(source_code_from_URL.text, 'html.parser')
    texts = bs.select('article > div > h4 > a')
    for text in texts:
        output_file.write(text.getText())
        
output_file.close() # 파일은 항상 열었으면 꼭 닫아야함    
```

```python
# 텍스트 파일에 저장된 단어로 워드 클라우드 
from konlpy.tag import Kkma
kkma = Kkma()

open_text_file = open(keyword + '.txt', 'r', encoding = 'utf8')
text = open_text_file.read()

# 명사만 추출 - 영문의 경우는 대부분은 공백을 기본으로 tokenizer 를 하고 불용어를 제거
# 한글의 경우는 형태소 분석을 해서 명사만 추출하고 불용어를 제거
nouns = kkma.nouns(text)
print(nouns)
open_text_file.close()
```

```python
# 단어의 개수 확인
import nltk
ko = nltk.Text(nouns, name = '박문석')
print(len(ko.tokens)) # 전체 단어의 수
print(len(set(ko.tokens))) # 중복을 제거한 단어의 수
print(ko.vocab()) # 단어별 등장 횟수
```

```python
# wordcloud 그리기
# 150개만 추출
data = ko.vocab().most_common(150)
wordcloud = WordCloud(font_path = 'data_nlp/NanumBarunGothic.ttf',
                     background_color = 'white').generate_from_frequencies(dict(data))
plt.imshow(wordcloud)
plt.show()
```

# 감성 분석

### 나이브 베이즈 분류기를 이용한 영문 감성 분석

```python
from nltk.tokenize import word_tokenize
import nltk

train = [
    ('I like you', 'pos'),
    ('I do not like you', 'neg'),
    ('I hate you', 'neg'),
    ('I do not hate you', 'pos'),
    ('I love you', 'pos'),
]

# 단어의 집합을 생성
# train 데이터의 모든 것을 순차적으로 sentence 에 대입한 후
# sentence의 첫 번째 데이터를 공백 기준으로 분할해서 word 에 순차적으로 할당한 수
# word를 소문자로 변경해서 set으로 생성
all_words = set(word.lower() 
                for sentence in train
                   for word in word_tokenize(sentence[0]))
print(all_words)
```

```python
# train 의 모든 데이터를 x 에 대입한 후
# 튜플을 생성하는데 뒤는 감성을 그대로 저장하고
# 앞은 all_words 모든 단어와 존재 여부를 딕셔너리로 가지고 있습니다.
t = [({word : (word in word_tokenize(x[0])) 
       for word in all_words}, x[1])
            for x in train]
print(t)
```

```python
# 나이블 베이즈 분류기를 이용해서 감성 분석을 하기 위해서는
# 각 문장이 단어의 존재 여부를 가진 dict와
# 감성을 나타내는 텍스트나 숫자의 튜플로 변환이 되어야 합니다.
```

```python
# 나이브 베이즈 분류기로 훈련
classifier = nltk.NaiveBayesClassifier.train(t)
# 각 단어가 감성에 미치는 확률
classifier.show_most_informative_features()
```

```python
# 예측 - 예측할 때 문장을 바로 대입하는 것이 아니고 문장을 피처로 변환해서 대입
test_sentence = 'I do not like Sue'

test_sent_features = {word.lower() : (word in word_tokenize(test_sentence.lower()))
                     for word in all_words}

# 테스트 문장을 피처화
print(test_sent_features)
```

```python
print(classifier.classify(test_sent_features))
# 이런 형태의 애플리케이션을 만들 때는 새로 들어오는 데이터를 가지고 
# 자동으로 학습하는 형태로 애플리케이션을 구성해야합니다.
```

### 나이브 베이즈를 이용한 한글 감성 분석

```python
# 샘플 데이터 생성
train = [('나는 당신을 사랑합니다.', '긍정'),
         ('나는 당신을 사랑하지 않습니다.', '부정'),
         ('나는 당신을 좋아합니다.', '긍정'),
         ('나는 당신을 만나는 것이 지루합니다.', '부정'),
         ('나는 노는 것이 좋습니다.', '긍정'),
        ]
```

```python
from nltk.tokenize import word_tokenize
import nltk

# 한글을 공백 기준으로 토큰화
all_words = set(word.lower() for sentence in train 
                for word in word_tokenize(sentence[0]))
# 동일한 어근을 갖는 '사랑하지'와 '사랑합니다'가 다른 단어로 구분
# 한글은 영문에서 사용하는 토크나이저를 쓸 수 없습니다.
# 형태소 분석기를 사용해야합니다.
print(all_words)
```

```python
# 한글 형태소 분석 - 단어 단위로 분할하고 품사까지 묶기
# 다른 형태소 분석기를 사용해도 되는데 품사를 같이 리턴하는 함수를 이용하는 것이 좋습니다.
# 품사를 제외하고 만든 분서기 보다 품사를 포함시키는 것이 성능이 우수하다고 알려져 있습니다.
# 감성분석에서 주로 사용 Twitter (품사를 포함)
from konlpy.tag import Twitter
twitter = Twitter()

# 형태소 분석을 해서 단어와 품사 사이에 /를 추가해서 구분해주는 함수
def tokenize(doc) :
    return["/".join(t) for t in twitter.pos(doc, norm = True, stem = True)]

# 훈련 데이터의 문장을 하나씩 함수에 대입하고 결과를 list로 묶기
train_docs = [(tokenize(row[0]), row[1]) for row in train]
print(train_docs)
```

```python
# 감성을 제외하고 단어만 추출해서 list로 만들기
tokens = [ t for d in train_docs for t in  d[0]]
print(tokens)
```

```python
# 나이브 베이즈을 이용한 감성 분석에서 사용할 단어 사전을 생성
def term_exists(doc) :
    return {word : (word in set(doc)) for word in tokens}

# 각 문장들의 단어의 포함 여부를 확인하고 긍정인지 부정인지 값을 설정
# 실제 작업에서 이 과정과 다음 과정이 시간이 많이 걸립니다.
train_xy = [(term_exists(d), c) for d, c in train_docs]
print(train_xy)
```

```python
# 나이브 베이즈 감성 분류기 생성 : 단어의 존재 여부에 따라 pos 와 neg 의 확률을 구합니다.
classifier = nltk.NaiveBayesClassifier.train(train_xy)
classifier.show_most_informative_features()
```

```python
# 예측
test_sentence = [('나는 당신을 만나는 것을 좋아합니다')]

# 형태소 분석
test_docs = twitter.pos(test_sentence[0])
print(test_docs)
print('========================================================================')
# 단어 사전에 있는 단어의 존재 여부를 확인
test_sent_features = {word : (word in tokens) for word in test_docs}
print(test_sent_features)
```

```python
# 예측
print(classifier.classify(test_sent_features))
```

### Kaggle의 IMDB 영화평을 이용한 지도 학습 기반의 감성 분석

```python
# 데이터 읽어오기
# tsv는 탭으로 구분된 문자열

import pandas as pd

review_df = pd.read_csv("data_nlp/word2vec-nlp-tutorial/labeledTrainData.tsv/labeledTrainData.tsv", sep = "\t")
print(review_df)
# id는 리뷰를 작성한 유저의 아이디
# sentiment가 감성. 1이면 긍정, 0이면 부정
# review가 리뷰
```

```python
# 데이터 전처리

# <br /> 제거: 공백으로 치환
review_df['review'] = review_df['review'].str.replace("<br />", " ")

import re
# 영문자로 시작하지 않는 단어를 모두 제거
# a-ZA-Z가 아닌 글자를 공백으로 치환
# 한글을 제외한 글자 제거: [^가-힣]
review_df['review'] = review_df['review'].apply(
    lambda x : re.sub("[^a-zA-Z]", " ", x))

print(review_df['review'].head())
```

```python
# 훈련 데이터와 검증 데이터 분리
from sklearn.model_selection import train_test_split

class_df = review_df['sentiment']
feature_df = review_df.drop(['id', 'sentiment'], axis = 1, inplace = False)

X_train, X_test, y_train, y_test = train_test_split(feature_df, class_df,
                                                   test_size = 0.3,
                                                   random_state = 42)
print(X_train.shape)
print(X_test.shape)
```

```python
from sklearn.feature_extraction.text import CountVectorizer, TfidfVectorizer
from sklearn.pipeline import Pipeline
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score, roc_auc_score
```

```python
# CountVectorizer를 이용한 피처 벡터화 후 분류 모델 훈련
pipeline = Pipeline([
    ('cnt_vect', CountVectorizer(stop_words = 'english', ngram_range = (1, 2))),
    ('lr_clf', LogisticRegression(C = 10))
])

pipeline.fit(X_train['review'], y_train)
pred = pipeline.predict(X_test['review'])
acc = accuracy_score(y_test, pred)

# ROC_AUC는 예측한 클래스 값을 이용하는 것이 아니라 확률을 이용
pred_probs = pipeline.predict_proba(X_test['review'])[:, 1]
roc = roc_auc_score(y_test, pred_probs)
print("정확도:", acc)
print("ROC_AUC:", roc)
```

```python
# TfidfVectorizer를 이용한 피처 벡터화 후 분류 모델 훈련
pipeline = Pipeline([
    ('tfidf_vect', TfidfVectorizer(stop_words = 'english', ngram_range = (1, 2))),
    ('lr_clf', LogisticRegression(C = 10))
])

pipeline.fit(X_train['review'], y_train)
pred = pipeline.predict(X_test['review'])
acc = accuracy_score(y_test, pred)

# ROC_AUC는 예측한 클래스 값을 이용하는 것이 아니라 확률을 이용
pred_probs = pipeline.predict_proba(X_test['review'])[:, 1]
roc = roc_auc_score(y_test, pred_probs)
print("정확도:", acc)
print("ROC_AUC:", roc)
```

### 네이버 식당 리뷰를 이용한 감성 분석

```python
# 데이터 가져오기
df = pd.read_csv("data_ML/review_data.csv")
df.info()
```

```python
# 한글을 제외한 글자 전부 제거해주는 함수
def text_cleaning(text):
    hangul = re.compile('[^ㄱ-ㅣ 가-힣]')
    result = hangul.sub('', text)
    return result
```

```python
df['ko_text'] = df['review'].apply(lambda x: text_cleaning(x))
del df['review'] # del: 메모리 정리
df.head()
```

```python
# 형태소 분석
from konlpy.tag import Okt

# 문장을 받아서 형태소 분석을 수행해 결과를 리턴해주는 함수
# 단어와 품사를 /로 구분해서 리턴
def get_pos(x):
    tagger = Okt()
    pos = tagger.pos(x)
    pos = ['{}/{}'.format(word, tag) for word, tag in pos]
    return pos
```

```python
result = get_pos(df['ko_text'][0])
print(result)
```

```python
# 등장 횟수만을 고려한 피처 벡터화를 위한 사전 생성을 생성하고 데이터를 피처 벡터화 수행
from sklearn.feature_extraction.text import CountVectorizer

index_vectorizer = CountVectorizer(tokenizer = lambda x: get_pos(x))
X = index_vectorizer.fit_transform(df['ko_text'].tolist())
print(X.shape)
# 3030이 단어의 개수
# 545는 행의 개수
```

```python
print(df['ko_text'][0])
print(X[0])

# 얘네보다 Tfidf가 나을 때가 있음
```

```python
# CountVectorizer를 이용해서 만든 피처 벡터를 Tfidf 기반으로 변경
from sklearn.feature_extraction.text import TfidfTransformer

tfidf_vectorizer = TfidfTransformer()
X = tfidf_vectorizer.fit_transform(X)
print(X[0])
```

```python
# 훈련 데이터와 검증 데이터 생성
from sklearn.model_selection import train_test_split

y = df['y']
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.3, random_state = 42)

print(X_train.shape)
print(X_test.shape)
```

```python
# 모델 생성 및 훈련과 평가
from sklearn.linear_model import LogisticRegression

from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score

lr = LogisticRegression(random_state = 42)
lr.fit(X_train, y_train)
y_pred = lr.predict(X_test)
y_pred_probability = lr.predict_proba(X_test)

print("정확도:", accuracy_score(y_test, y_pred))
print("precision:", precision_score(y_test, y_pred))
print("recall:", recall_score(y_test, y_pred)) # 1  나옴. 이런 건 의심해 볼 필요 있음!
print("F1:", f1_score(y_test, y_pred))
```

```python
from sklearn.metrics import confusion_matrix

# 모든 데이터를 1로 분류 => 이래서 정확도가 높게 나온 거!
# 실제 1인 데이터는 전부 1로 분류했으므로 recall이 1
print(confusion_matrix(y_test, y_pred))
```

## 문서 군집

```python
# 읽어야 하는 파일의 목록 만들기
import glob, os
path = "data_nlp/OpinosisDataset/topics/" # 문서가 들어있는 디렉토리 경로
# path에 있는 확장자가 data인 모든 파일의 경로를 저장
all_files = glob.glob(os.path.join(path, "*.data"))
print(all_files)
```

```python
#데이터를 읽어서 저장

#파일의 이름을 저장할 list
filename_list = []
#텍스트를 저장할 list
opinion_text = []

#모든 파일의 내용을 읽어서 저장
for file_ in all_files:
    #latin1은 서유럽 인코딩 - 한글이 안됩니다.
    df = pd.read_table(file_, index_col = None, header = 0, encoding='latin1')
    print(df)
    
    #파일 이름만 저장
    filename = file_.split("\\")[-1] #디렉토리 기호로 구분해서 마지막 부분을 저장
    filename = filename.split(".")[0] # 확장자 제거
    filename_list.append(filename)
    
    #내용 저장
    opinion_text.append(df.to_string())

print(filename_list)
print(opinion_text)
```

```python
document_df = pd.DataFrame({'filename' : filename_list, 'opinion_text' : opinion_text})
document_df.head()
```

```python
# 영문 품사 태깅
from nltk.stem import WordNetLemmatizer
import nltk
import string

# 구두점 관련 디셔너리를 생성
# 구두점을 None으로 치환하기 위한 딕셔너리
# 앞의 숫자는 ASCII Code 값
remove_punct_dict = dict((ord(punct), None) for punct in string.punctuation)
# print(remove_punct_dict)

lemmar = WordNetLemmatizer()

# 품사 태깅해주는 함수
def LemTokens(tokens) :
    return[lemmar.lemmatize(token) for token in tokens]

# 품사 태깅과 구두점을 제거하는 함수
def LemNormalize(text) :
    return LemTokens(nltk.word_tokenize(text.lower().translate(remove_punct_dict)))

from sklearn.feature_extraction.text import TfidfVectorizer

tfidf_vect = TfidfVectorizer(tokenizer = LemNormalize, stop_words = 'english',
                            ngram_range = (1, 2), min_df = 0.06, max_df = 0.85)

feature_vect = tfidf_vect.fit_transform(document_df['opinion_text'])
```

```python
feature_vect
```

```python
from sklearn.cluster import KMeans

km_cluster = KMeans(n_cluster = 5, max_iter = 10000, random_state = 42)
km_cluster.fit(feature_vect)
cluster_label = km_cluster.labels_
cluster_centers = km_cluster.cluster_centers_
```

```python
document_df['cluster_label'] = cluster_label
document_df.head()
```
